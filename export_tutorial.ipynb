# Exporting Data in Flat and SQL Formats

This tutorial demonstrates how to export data in flat CSV and SQL formats using the `mario` library. We will use an Input-Output Table (IOT) as an example.

## Step 1: Importing the IOT Table

First, we import the IOT table using the `mario.parse_exiobase` function. Make sure to specify the correct path to your dataset.

```python
import mario

table = 'IOT'

if table == 'IOT':
    db = mario.parse_exiobase(
        path='/Users/lorenzorinaldi/Library/CloudStorage/OneDrive-SharedLibraries-PolitecnicodiMilano/DENG-SESAM - Documenti/c-Research/a-Datasets/Exiobase/Monetary/v3.8.2/IOT/IOT_2022_ixi.zip',
        table=table,
        unit="Monetary",
        version='3.8.2',
    )

elif table == 'SUT':
    db = mario.parse_from_txt(
        path = '/Users/lorenzorinaldi/Library/CloudStorage/OneDrive-SharedLibraries-PolitecnicodiMilano/DENG-SESAM - Documenti/c-Research/a-Datasets/Exiobase/Hybrid/3.3.18 with VA/flows',
        table = table,
        mode = 'flows',
    )

elif table == 'test_IOT':
    table = 'IOT'
    db = mario.load_test(table)
else:
    table = 'SUT'
    db = mario.load_test(table)
```

## Step 2: Aggregating the Data

Aggregate the IOT data using the `aggregate` method. Provide the name of the Excel file where the aggregated data will be saved.

```python
db.aggregate(f"aggregate_{table}.xlsx")
```

## Step 3: Cloning a Scenario

Create a clone of the baseline scenario to allow exporting with multiple scenarios.

```python
db.clone_scenario("baseline","Test - 2022")
```

## Step 4: Exporting to Flat CSV Format

Export the data to a flat CSV format using the `to_flat_csv` method. You can customize the export options, such as including flows and coefficients, and splitting scenarios into multiple columns.

```python
db.to_flat_txt(
    path=f'Test_export_{table}',
    matrices = ['f'],
    scenario_split={
        "separator": " - ",
        "Scenario": 0,
        "Year": 1,
        "rename_baseline": "Baseline - 2022",
    },
)
```

## Step 5: Exporting to SQL Format

Export the data to an SQL database using the `to_sql` method. The arguments are similar to those used for exporting to CSV.

```python
db.to_sql(
    path=f'Test_export_{table}',
    matrices = ['f'],
    scenario_split={
        "separator": " - ",
        "Scenario": 0,
        "Year": 1,
        "rename_baseline": "Baseline - 2022",
    },
)
```

## Conclusion

You have now learned how to export data in flat CSV and SQL formats using the `mario` library. Customize the paths and options as needed for your specific use case. 
The very same procedure works seamlessly with Supply and Use Tables (SUTs) as well.
