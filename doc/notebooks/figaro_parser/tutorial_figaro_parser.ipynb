# FIGARO parser

FIGARO is parsed directly from the Eurostat API. A separate download step is not required.

```python
import mario
```

Use `countries=` while testing workflows, then remove it to parse the full FIGARO table.

```python
db = mario.parse_figaro(
    table="SUT",
    year=2022,
    countries=["IT", "ME"],
    calc_all=False,
)

db
```

```python
db_iot = mario.parse_figaro(
    table="IOT",
    year=2022,
    iot_mode="product",
    countries=["IT", "ME"],
    calc_all=False,
)

db_iot
```
