# Large database workflows

This notebook uses the packaged IOT test table to show the same habits that matter on large databases: calculate only what you need and prefer solve-based calculations for demand-driven results.

```python
import mario

mario.set_log_verbosity("critical")
```

```python
db = mario.load_test("IOT")
```

## Start from the stored matrices

```python
db.list_matrices()
```

## Calculate only selected matrices

```python
db.calc_all(
    ["X", "z"],
    compute_method="solve",
    linear_solver="scipy",
)
db.list_matrices()
```

## Resolve a second small group of matrices

```python
resolved = db.resolve_many(
    ["f", "m"],
    compute_method="solve",
    linear_solver="scipy",
)
resolved.keys()
```

## Use `query` to inspect only the result you need

```python
db.query("f")
```
