# Change settings

This notebook changes MARIO compute settings for a small calculation, then restores the previous settings.

## Warning: settings changes persist

> Helpers such as `mario.set_compute_method(...)`, `mario.set_linear_solver(...)`, and `mario.upload_settings(...)` update MARIO's active settings file for the current installation or environment.

> That means the new defaults persist across future Python sessions and future `import mario` calls until you change them again or call `mario.reset_settings()`.

> This notebook therefore saves the current settings first and restores them explicitly at the end.

```python
import mario

mario.set_log_verbosity("critical")
```

## Store the current settings

```python
previous_settings = mario.download_settings(None)
previous_settings["compute"]
```

## Change the compute settings

```python
mario.set_compute_method("solve")
mario.set_linear_solver("scipy")
mario.set_linear_strategy("auto")

mario.download_settings(None)["compute"]
```

## Run a calculation with the new defaults

```python
db = mario.load_test("IOT")
db.calc_all(["f"])
db.f
```

## Restore the previous settings

```python
mario.upload_settings(previous_settings)
mario.download_settings(None)["compute"]
```
