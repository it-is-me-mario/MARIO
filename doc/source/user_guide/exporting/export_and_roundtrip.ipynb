# Exporting

This notebook shows a minimal export workflow starting from the packaged IOT test table.
It writes the same database to Excel, CSV, and Parquet so you can inspect the output layouts and reuse them in downstream workflows.

```python
import mario
db = mario.load_test("IOT")
```

## Excel export

`to_excel()` writes a workbook that can include both flow matrices and coefficient matrices in one file.
This is usually the most convenient format when you want to inspect the exported structure manually in Excel.

```python
db.to_excel(
    path="/path/to/iot_export.xlsx",
    flows=True,
    coefficients=True,
)
```

The exported workbook keeps the familiar matrix-style layout, with separate sheets for the main blocks that MARIO writes.

![export_excel](../../_static/images/export_excel.png)

## CSV and Parquet export

For larger datasets, directory-based exports are often easier to version, inspect, and process outside Excel.
With `flat=True`, MARIO writes one compact `data` file plus a `units` file, which is the layout expected by the flat TXT/CSV and Parquet parsers.

```python
db.to_txt(
    path="/path/to/iot_export_csv",
    flows=True,
    _format="csv",
)
```

```python
db.to_txt(
    path="/path/to/iot_export_csv",
    flows=False,
    coefficients=True,
    _format="csv",
    flat=True,
)
```

```python
db.to_parquet(
    path="/path/to/iot_export_parquet",
    flows=True,
    coefficients=True,
    flat=True,
)
```

The flat export layout is more compact than the Excel workbook and is usually the better choice for roundtrip workflows and larger filesystems.

![export_csv](../../_static/images/export_csv.png)

The files written here can be parsed back with the custom TXT/CSV and Parquet parsers described in the parser section of the documentation.
