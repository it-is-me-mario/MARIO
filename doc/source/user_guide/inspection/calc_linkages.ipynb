# Calculate linkages

This notebook calculates backward and forward linkages from the packaged IOT test table.

In practice, linkages answer two different questions:

- **Backward linkage**: if final demand for one sector increases, how strongly does that sector pull activity from its upstream suppliers? High backward linkage usually points to sectors that mobilize many inputs along the supply chain.
- **Forward linkage**: if output from one sector changes, how strongly does that propagate into downstream sectors that use it as an input? High forward linkage usually points to foundational suppliers, materials, or enabling sectors.

A simple practical reading is often:

- high **backward** = a strong demander of upstream inputs
- high **forward** = a strong provider to downstream sectors
- high **both** = a structurally important node in the production network

MARIO now exposes two extra low-cost diagnostics alongside the standard linkage measures:

- In aggregated mode (``multi_mode=False``), it adds ``Forward Amplification`` and ``Backward Amplification`` ratios computed as ``Total / Direct``.
- In multi-regional mode (``multi_mode=True``), it adds ``Local Share`` and ``Foreign Share`` next to the existing ``Local`` and ``Foreign`` linkage components.

```python
import mario

mario.set_log_verbosity("critical")
```

```python
db = mario.load_test("IOT")
```

## Calculate standard linkages

```python
linkages = db.calc_linkages()
linkages
```

The default multi-regional output keeps the absolute ``Local`` and ``Foreign`` components for each linkage and now also reports their shares. This makes it easier to distinguish the size of a linkage from its domestic-versus-external composition.

A practical way to read the columns is:

- ``Direct``: the first-round effect only
- ``Total``: direct plus all indirect rounds through the production network
- ``Local`` / ``Foreign``: where that linkage is geographically activated
- ``Local Share`` / ``Foreign Share``: how the linkage is split, regardless of absolute size

So two sectors can have the same total linkage but very different supply-chain geography: one mostly domestic, the other heavily dependent on foreign production chains.

## Calculate aggregated linkages and amplification ratios

```python
aggregated_linkages = db.calc_linkages(multi_mode=False)
aggregated_linkages
```

``Forward Amplification`` and ``Backward Amplification`` are computed as ``Total / Direct``. Values above 1 indicate that indirect supply-chain effects amplify the direct linkage.

In practical terms, a sector with a modest direct linkage but a high amplification ratio is one where the first-round effect looks limited, yet the wider chain reaction through suppliers or users is much larger. This is often the interesting case when you want to identify hidden systemic importance rather than only the obvious first-tier effect.

## Calculate non-normalized linkages

```python
db.calc_linkages(normalized=False)
```

With ``normalized=False``, MARIO returns raw linkage magnitudes instead of relative scores around the system average.

Use the normalized view when the goal is comparison across sectors. Use the non-normalized view when you care about absolute magnitude and want to preserve scale instead of asking whether a sector is above or below average.
