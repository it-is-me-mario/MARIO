# Provide your database

This notebook creates a MARIO-readable Excel template and parses it back. The labels mirror the packaged MARIO test tables.

```python
import mario

mario.set_log_verbosity("critical")
```

## Write an IOT template

```python
mario.write_parse_template(
    "custom_iot_template.xlsx",
    table="IOT",
    sets={
        "regions": ["Reg1", "Reg2"],
        "sectors": ["Agriculture", "Services"],
        "final demand": ["Final demand"],
        "factors": ["Taxes", "Wages"],
        "satellites": ["CO2"],
    },
    units={
        "sectors": "M EUR",
        "factors": "M EUR",
        "satellites": "kg",
    },
)
```

## Parse the template back

The numeric cells are empty, so MARIO reads them as zeros. This is enough to validate the structure before filling real data.

```python
db = mario.parse_from_excel(
    path="custom_iot_template.xlsx",
    table="IOT",
    mode="flows",
)
db.get_index("all")
```

## Write a SUT template

```python
mario.write_parse_template(
    "custom_sut_template.xlsx",
    table="SUT",
    sets={
        "regions": ["Region 1"],
        "activities": ["Manufacturing", "Services"],
        "commodities": ["Goods", "Services"],
        "final demand": ["Final demand"],
        "factors": ["Taxes"],
        "satellites": ["CO2"],
    },
    units={
        "activities": "M EUR",
        "commodities": "M EUR",
        "factors": "M EUR",
        "satellites": "kg",
    },
)

sut = mario.parse_from_excel(
    path="custom_sut_template.xlsx",
    table="SUT",
    mode="flows",
)
sut.get_index("all")
```
