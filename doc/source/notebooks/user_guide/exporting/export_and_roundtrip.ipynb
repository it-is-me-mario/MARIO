# Export and roundtrip

This notebook exports the packaged IOT test table and parses the exported files back into MARIO.

```python
import mario

mario.set_log_verbosity("critical")
```

```python
db = mario.load_test("IOT")
db.calc_all(["z", "v", "e"])
```

## Excel roundtrip

```python
db.to_excel(
    path="iot_export.xlsx",
    flows=True,
    coefficients=True,
)

excel_flows = mario.parse_from_excel(
    path="iot_export.xlsx",
    table="IOT",
    mode="flows",
)
excel_flows.Z.shape
```

## CSV roundtrip

The flat export writes one compact `data` file plus a `units` file.

```python
db.to_txt(
    path="roundtrip_csv",
    flows=True,
    coefficients=True,
    _format="csv",
    flat=True,
)

csv_flows = mario.parse_from_txt(
    path="roundtrip_csv/flows",
    table="IOT",
    mode="flows",
    _format="csv",
    flat=True,
)
csv_flows.Z.shape
```

## Parquet roundtrip

```python
db.to_parquet(
    path="roundtrip_parquet",
    flows=True,
    coefficients=True,
    flat=True,
)

parquet_flows = mario.parse_from_parquet(
    path="roundtrip_parquet/flows",
    table="IOT",
    mode="flows",
    flat=True,
)
parquet_flows.Z.shape
```
