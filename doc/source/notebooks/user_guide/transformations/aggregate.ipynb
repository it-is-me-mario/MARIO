# Aggregate

This notebook uses the packaged IOT test table and an Excel aggregation workbook.

```python
import mario

mario.set_log_verbosity("critical")
```

```python
db = mario.load_test("IOT")
```

## Generate an empty Excel template

Open the generated workbook if you want to see the format MARIO expects.

```python
db.get_aggregation_excel(
    path="aggregation_template.xlsx",
    levels=["Region", "Sector"],
    overwrite=True,
)
```

## Apply a filled Excel workbook

The example workbook groups `Reg1` as `Domestic`, `Reg2` as `External`, and sectors into two groups.

```python
aggregated = db.aggregate(
    io="../excel/aggregation_iot.xlsx",
    levels=["Region", "Sector"],
    inplace=False,
)
aggregated.get_index("all")
```
