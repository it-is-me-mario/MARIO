# Apply shocks

This notebook creates a new scenario from an Excel shock workbook using the packaged IOT test table.

```python
import mario

mario.set_log_verbosity("critical")
```

```python
db = mario.load_test("IOT")
```

## Generate an empty shock template

```python
db.get_shock_excel(
    path="shock_template.xlsx",
    num_shock=2,
)
```

## Apply a filled shock workbook

The filled workbook updates one `z` coefficient.

```python
db.shock_calc(
    io="../excel/shock_z_iot.xlsx",
    z=True,
    scenario="example shock",
)
db.scenarios
```

## Inspect the shocked coefficient matrix

```python
db.query("z", scenarios=["example shock"]).head()
```
