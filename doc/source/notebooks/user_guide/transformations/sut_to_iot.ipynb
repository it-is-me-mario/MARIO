# SUT to IOT

This notebook transforms the packaged SUT test table into an IOT.

```python
import mario

mario.set_log_verbosity("critical")
```

```python
sut = mario.load_test("SUT")
```

## Inspect the starting SUT

```python
sut.table_type
```

```python
sut.get_index("all")
```

## Transform to IOT

```python
iot = sut.to_iot(method="A", inplace=False)
iot.table_type
```

## Inspect the resulting IOT

```python
iot.get_index("all")
```

```python
iot.Z
```
