# Add sectors

This notebook adds one sector to the packaged IOT test table using an Excel add-sector workbook.

```python
import mario

mario.set_log_verbosity("critical")
```

```python
db = mario.load_test("IOT")
```

## Generate an empty add-sector template

```python
db.get_add_sectors_excel(
    items=["New sector"],
    regions=["Reg1"],
    path="add_sector_template.xlsx",
    overwrite=True,
)
```

## Apply a filled add-sector workbook

```python
expanded = db.add_sectors(
    io="../excel/add_sector_iot.xlsx",
    inplace=False,
)
expanded.get_index("Sector")
```

## Check the new sector coefficient

```python
expanded.z.loc[("Reg1", "Sector", "Agriculture"), ("Reg1", "Sector", "New sector")]
```
