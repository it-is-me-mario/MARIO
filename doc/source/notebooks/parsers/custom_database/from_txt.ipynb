# TXT, CSV, and Parquet custom parser walkthrough

This walkthrough covers the directory-based parsers for TXT, CSV, and Parquet payloads.


## Main Entry Points

- `mario.parse_from_txt(...)`
- `mario.parse_from_parquet(...)`


```python
import mario
```

## Matrix-per-file TXT or CSV

Use `flat=False` for the historical matrix-per-file layout.


```python
db = mario.parse_from_txt(
    path="/Users/lorenzorinaldi/Documents/GitHub/MARIO/mario/test/supporting_files/iot_export_csv/flows",
    table="IOT",
    mode="flows",
    _format="csv",
    flat=False,
)
```

## Flat TXT or CSV

Use `flat=True` for long-format payloads. MARIO accepts either one combined `data` file or separate flat files per matrix, as long as `units` is present.


```python
db = mario.parse_from_txt(
    path="/Users/lorenzorinaldi/Documents/GitHub/MARIO/mario/test/supporting_files/iot_export_csv/coefficients",
    table="IOT",
    mode="coefficients",
    _format="csv",  
    flat=True,
)
```

## Flat Parquet

The same logic applies to Parquet exports.


```python
db = mario.parse_from_parquet(
    path="/Users/lorenzorinaldi/Documents/GitHub/MARIO/mario/test/supporting_files/iot_export_parquet/flows",
    table="IOT",
    mode="flows",
    flat=True,
)
```
