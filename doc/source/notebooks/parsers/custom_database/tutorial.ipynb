# Custom Databases

This notebook mirrors the custom-database parser guide.


## Main entry points

- `mario.parse_from_excel(...)`
- `mario.parse_from_txt(...)`
- `mario.parse_from_parquet(...)`


```python
import mario
```

## Parse a custom Excel IOT


```python
db = mario.parse_from_excel(
    path="/path/to/custom_iot.xlsx",
    table="IOT",
    mode="flows",
)
```

## Parse a custom Excel SUT with product-based assumption


```python
db = mario.parse_from_excel(
    path="/path/to/custom_sut.xlsx",
    table="SUT",
    mode="flows",
    tech_assumption="PT",
)
```

## Parse flat TXT or Parquet exports


```python
db = mario.parse_from_txt(
    path="/path/to/txt_directory",
    table="IOT",
    mode="flows",
    flat=True,
)
```

```python
db = mario.parse_from_parquet(
    path="/path/to/parquet_directory",
    table="SUT",
    mode="coefficients",
    flat=True,
)
```
