# StatCan

This notebook mirrors the StatCan parser guide.


## Main entry point

Use `mario.parse_statcan(...)` for StatCan WDS workflows.


```python
import mario
```

## Parse a SUT directly from WDS


```python
db = mario.parse_statcan(
    year=2022,
    table="SUT",
    level="summary",
    geo="Canada",
)
```

## Parse an IOT


```python
db = mario.parse_statcan(
    year=2022,
    table="IOT",
    level="detail",
    valuation="basic",
)
```

## Keep the raw CSV locally


```python
db = mario.parse_statcan(
    year=2022,
    table="SUT",
    level="detail",
    path="/path/to/statcan_cache",
    download=True,
)
```
