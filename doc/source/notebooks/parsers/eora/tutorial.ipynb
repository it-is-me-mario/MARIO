# EORA

This notebook mirrors the EORA parser guide.


## Main entry point

Use `mario.parse_eora(...)` for both EORA26 multi-region and single-region workflows.


## Key arguments

The key public arguments are:

- `path`: one EORA file or one dataset directory
- `multi_region`: `True` for EORA26, `False` for the single-region workflow
- `table`: mainly relevant for multi-region parsing, where the workflow is `IOT`
- `indeces`: optional path to the EORA26 label files
- `name_convention`: choose abbreviations or full country names
- `aggregate_trade`: single-region helper to aggregate trade rows
- `country`: selects one country file in the single-region workflow


```python
import mario
```

## Parse EORA26


```python
db = mario.parse_eora(
    path="/path/to/eora26_directory",
    multi_region=True,
    table="IOT",
    indeces="/path/to/eora26_labels",
)
```

## Parse a single-region EORA file


```python
db = mario.parse_eora(
    path="/path/to/single_region_directory",
    multi_region=False,
    country="ITA",
)
```
