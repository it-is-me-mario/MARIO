# ADB

This notebook mirrors the practical ADB parser walkthrough in the documentation.


## Main entry point

Use `mario.parse_adb(...)` for the public workflow.


## Source

- Official ADB MRIO and SRIO page: [ADB globalization portal](https://kidb.adb.org/globalization/current)
- Official ADB air-emissions extensions page: [ADB environmentally extended MRIOT](https://kidb.adb.org/globalization/adb_environmentally_extended_multiregional_inputoutput_tables)


## Key arguments

The key public arguments are:

- `path`: one ADB workbook or one directory containing multiple workbooks
- `table`: currently only `"IOT"` is supported
- `year`: use it when one directory contains more than one MRIO release year; for SRIO workbooks it is mandatory because one file contains multiple yearly sheets
- `economies`: disambiguates multiple workbook variants for the same year
- `add_extensions`: optional path to the ADB air-emissions workbook; this populates `E` and keeps `EY` zero-filled


```python
import mario
```

## Parse one workbook


```python
db = mario.parse_adb(
    path="/path/to/ADB_MRIO_2024_74.xlsx",
)
```

## Parse from a directory


```python
db = mario.parse_adb(
    path="/path/to/adb_directory",
    year=2024,
    economies=74,
)
```

## Parse one SRIO workbook

For ADB SRIO workbooks, `year=` is required because the workbook contains one sheet per year.


```python
db = mario.parse_adb(
    path="/path/to/CAN IOT 2000, 2007-2024.xlsx",
    year=2024,
)
```

## Add air-emissions extensions

The same `add_extensions=` workflow works for both MRIO and SRIO economic tables.


```python
db = mario.parse_adb(
    path="/path/to/ADB_MRIO_2024_74.xlsx",
    add_extensions="/path/to/2024 EE-MRIOT (Air Emissions).xlsx",
)
```

```python
db = mario.parse_adb(
    path="/path/to/CAN IOT 2000, 2007-2024.xlsx",
    year=2024,
    add_extensions="/path/to/2024 EE-MRIOT (Air Emissions).xlsx",
)
```
