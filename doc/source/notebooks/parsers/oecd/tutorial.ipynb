# OECD ICIO

This notebook mirrors the OECD parser guide.


## Main entry point

Use `mario.parse_oecd(...)` for the public OECD ICIO workflow.


## Source link

- [OECD ICIO dataset page](https://www.oecd.org/en/data/datasets/inter-country-input-output-tables.html)


## Key arguments

The key public arguments are:

- `path`: one yearly OECD ICIO csv file or one directory containing multiple yearly files
- `year`: use it when the directory contains more than one year
- `name`: optional metadata label override
- `calc_all`: optional eager computation of derived blocks after parsing


```python
import mario
```

## Parse one yearly CSV


```python
db = mario.parse_oecd(
    path="/path/to/ICIO2021_2018.csv",
)
```

## Parse from a directory


```python
db = mario.parse_oecd(
    path="/path/to/oecd_directory",
    year=2018,
)
```
