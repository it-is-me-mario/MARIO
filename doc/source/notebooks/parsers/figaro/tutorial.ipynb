## Key arguments

The key public arguments are:

- `path`: local directory containing the FIGARO flat files
- `table`: choose `"SUT"` or `"IOT"`
- `year`: use it when the directory contains more than one FIGARO year
- `iot_mode`: IOT-only selector, use `"auto"`, `"product"` or `"industry"`


## Local layout expectation

MARIO expects one local directory containing the relevant FIGARO flat files, either as:

- `.zip` bundles;
- extracted `.csv` files.

For `table="SUT"`, MARIO looks for the supply/use files.
For `table="IOT"`, MARIO looks for the pxp and/or ixi IOT files.


```python
import mario
```

## Parse a FIGARO SUT

```python
db_sut = mario.parse_figaro(
    path="/path/to/figaro_directory",
    table="SUT",
)
```

## Parse a FIGARO IOT

For IOT parsing, `iot_mode` can be:

- `"auto"`
- `"product"`
- `"industry"`

When both IOT variants are present and `iot_mode="auto"`, MARIO defaults to the product-by-product file.


```python
db_iot = mario.parse_figaro(
    path="/path/to/figaro_directory",
    table="IOT",
    iot_mode="auto",
)
```
