# Eurostat SDMX

This notebook mirrors the Eurostat parser guide.


## Main entry point

Use `mario.parse_eurostat(...)` for both SUT and IOT workflows.


```python
import mario
```

## Parse a SUT directly from SDMX


```python
db = mario.parse_eurostat(
    country="IT",
    year=2022,
    table="SUT",
)
```

## Parse an IOT


```python
db = mario.parse_eurostat(
    country="IT",
    year=2022,
    table="IOT",
    iot_mode="product",
)
```

## Optional raw download


```python
mario.download_eurostat(
    path="/path/to/eurostat_cache",
    country="IT",
    year=2022,
    table="SUT",
)
```
