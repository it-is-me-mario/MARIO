# Dataset core from a MARIO database

This tutorial shows the smallest useful path into the new MARIO core.
We start from a regular `mario.Database`, convert it into a `mario.Dataset`,
and then use the new compute and scenario interfaces.


## Goals

- load a built-in IOT example with the Database API;
- convert it to the new `Dataset` model;
- compute a derived matrix through the new resolver;
- create a derived scenario and override one block.


```python
import mario

db = mario.load_test("IOT")
dataset = mario.Dataset.from_database(db)
```

```python
print(type(db).__name__)
print(type(dataset).__name__)
print(dataset.table_kind.value)
print(dataset.list_scenarios())
```

```python
dataset.list_blocks()
```

## Compute a missing block

The block `w` is not parsed directly in the test dataset. The resolver computes it
from the blocks already available in the dataset and stores the result.


```python
dataset.has_block("w")
```

```python
w = dataset.compute("w")
w.shape
```

```python
dataset.has_block("w")
```

```python
print(dataset.explain("p"))
```

## Create a scenario override

Scenarios in the new model inherit from one another. Here we create a `policy`
scenario, override `X`, and compute `w` again in the new scenario.


```python
dataset.create_scenario("policy", parent="baseline")

new_x = dataset.get_block("X").copy() * 0.95
dataset.set_block("X", new_x, scenario="policy")

baseline_w = dataset.compute("w", scenario="baseline")
policy_w = dataset.compute("w", scenario="policy")

print(baseline_w.shape)
print(policy_w.shape)
```

This is the core pattern behind the new architecture: keep the domain grammar,
but separate storage, scenario inheritance and compute logic more clearly.
