# Scenarios, comparisons and export

This tutorial stays entirely on the `Database` API and focuses on
three very common tasks: clone a scenario, compare it to baseline, and export the
result.


```python
from pathlib import Path
from tempfile import TemporaryDirectory

import mario

db = mario.load_test("IOT")
```

## Create a scenario variation


```python
db.clone_scenario("baseline", "policy")
db.matrices["policy"]["Z"] = db.matrices["policy"]["Z"] * 1.01

print(db.scenarios)
```

## Compare the policy scenario with baseline


```python
relative_z = db.query(
    matrices=["Z"],
    scenarios=["policy"],
    base_scenario="baseline",
    type="relative",
)

relative_z.shape
```

## Export to `pymrio`

The classic high-level export method remains available on the database object.


```python
io = db.to_pymrio(
    satellite_account="Extensions",
    factor_of_production="Value_added",
)

type(io).__name__
```

## Export to Excel

In practice you would usually write to a project output folder. Here we use a
temporary directory just to keep the notebook self-contained.


```python
with TemporaryDirectory() as tmp:
    output_path = Path(tmp) / "database.xlsx"
    db.to_excel(path=str(output_path), scenario="baseline")
    print(output_path.exists())
```

This is still the core public style of MARIO: methods live on the database object,
and common workflows remain short and imperative.
