# Parse and transform with the historical API

This tutorial stays on the old parser and transformation surface:
`parse_from_excel(...)`, `calc_all(...)` and `to_iot(...)`.


## Locate the packaged SUT workbook


```python
from pathlib import Path

import mario


def find_repo_root(start: Path) -> Path:
    for candidate in (start, *start.parents):
        if (candidate / "mario" / "test" / "SUT.xlsx").exists():
            return candidate
    raise FileNotFoundError("Could not locate the MARIO repository root.")


REPO_ROOT = find_repo_root(Path.cwd())
SUT_PATH = REPO_ROOT / "mario" / "test" / "SUT.xlsx"
SUT_PATH
```

## Parse a SUT database


```python
sut = mario.parse_from_excel(
    path=str(SUT_PATH),
    table="SUT",
    mode="flows",
    name="Tutorial SUT",
)

print(type(sut).__name__)
print(sut.table_type)
print(sut.scenarios)
```

## Transform the SUT into an IOT

This is still one of the central MARIO workflows. The method name and style remain
the historical ones.


```python
iot = sut.to_iot(method="B", inplace=False)

print(iot.table_type)
print(iot.Z.shape)
print(iot.Y.shape)
```

The main design principle is: preserve the familiar domain-facing methods, even if
the internals gradually become more modular.
