# Database quickstart

This tutorial uses the historical `mario.Database` API.
It is the right starting point if you already know MARIO or if you want the most
stable user-facing workflow while the internal restructuring continues.


## Load a built-in example

MARIO ships with small IOT and SUT fixtures that are useful for testing and for
learning the package.


```python
import mario

db = mario.load_test("IOT")
```

```python
print(type(db).__name__)
print(db.table_type)
print(db.scenarios)
```

```python
print(db.get_index("Region"))
print(db.get_index("Sector")[:5])
```

## Compute missing matrices

The public method remains `calc_all(...)`, exactly in the historical style.
Internally, MARIO may use newer compute paths, but the user-facing entry point is
still the old one.


```python
db.calc_all(["w", "p"])
```

```python
w = db.query(matrices=["w"], scenarios=["baseline"])
p = db.query(matrices=["p"], scenarios=["baseline"])

print(type(w).__name__, w.shape)
print(type(p).__name__, p.shape)
```
